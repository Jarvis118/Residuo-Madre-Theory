# RM-03 – Spettro CMB e rumore residuo post-collasso

Notebook per il confronto tra spettro CMB osservato e rumore di fondo residuo previsto dalla RM.

## Obiettivo
Verificare se la componente residua del rumore di fondo può essere interpretata come effetto del collasso del Residuo Madre.

## Criterio di falsificazione
Assenza di correlazione significativa (p < 0.05) → **RM violata**.

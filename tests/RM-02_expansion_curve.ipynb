# RM-02 – Espansione cosmica derivata dal Residuo Madre

Notebook di test per la curva di espansione H(z) prevista dalla RM.

## Obiettivo
Confrontare H(z) ottenuto dal modello RM con i dati osservativi (es. SDSS, DESI).

## Criterio di falsificazione
Se la curva H(z)_RM differisce >10% dai valori osservati medi → **RM violata**.

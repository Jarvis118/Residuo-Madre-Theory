# RM-01 – Verifica densità-energia oscura / materia oscura

Questo notebook rappresenta un **template di test** per la Teoria del Residuo Madre (RM).

## Obiettivo
Confrontare il rapporto ΩΛ / Ωm previsto dalla RM con i valori osservativi Planck (2018) e dataset più recenti.

## Struttura
- Input: parametri cosmologici (H₀, ΩΛ, Ωm)
- Output: differenza relativa e condizione di coerenza.

## Criterio di falsificazione
Se |ΩΛ_RM / Ωm_RM − ΩΛ_obs / Ωm_obs| > 0.05 → **RM violata**.
Altrimenti → **RM coerente**.
